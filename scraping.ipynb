#  Scraping Proyek Analisis Sentimen Aplikasi BRIMO

```python
from google_play_scraper import app, reviews_all, Sort
```

### Scrapping Review Google Play BRIMO

```python
scrapreview = reviews_all(
    'id.co.bri.brimo',     
    lang='id',             
    country='id',          
    sort=Sort.MOST_RELEVANT,
    count=3000             
)
```

### simpan ke file .csv

```python
import csv
 
with open('hasil_scraping.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])
    for review in scrapreview:
        writer.writerow([review['content']])
```

```python
import pandas as pd

app_reviews_df = pd.DataFrame(scrapreview)
app_reviews_df.shape
app_reviews_df.head()
app_reviews_df.to_csv('hasil_scraping.csv', index=False)
```
